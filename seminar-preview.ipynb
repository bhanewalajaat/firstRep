# Seaborn
- python library for visualization of data.

Seaborn has following type of plotting functions:

## Visualizing statistical relationships
- Variables in a dataset are said to be related if one variable changes, then the other variables also changes.
- scatter plot and line plots are used to check whether variables are related.
- Seaborn’s relplot() function  used to plot scatter and line plot.

### Scatter plot
- It depicts the relationship between two variables. 
- It plots points, where each point is an observation in the dataset. 
- There are various ways to plot a scatteplot. 
- But most common approach is to use relplot() function and setting kind=’scatter’
- Additional dimensions can be view by assigning categorical variables to row, col, hue parameters. 



```python
import matplotlib
%matplotlib inline
import seaborn as sns
data=pd.DataFrame({'time':np.arange(500),
                   'value':np.random.randn(500).cumsum()   
})

sns.relplot(data=data,x='time',y='value')
```

### Line plot
- If one of the variable is function of time better to use line plot. 
- set kind=’line’ in relplot() function.

```python
sns.relplot(data=data,x='time',y='value',kind='line')
```

- This function accept x, y, and data as arguments.  
- x for x axis variable,y for y axis variable 
- data for dataframe on which plot is to drawn,

- Possible to view Additional dimensions

- set categorical variable to hue, row, col arguments.

### Categorical distribution plots
- As the size of datasets grows, categorical scatter plots become limited in the information they can provide. 
- several alternative approaches for summarization of categorical variables.


1. ####  Boxplots
    - shows three quartile values in the box, 
    - whiskers extends to the points that lie in the 1.5*QR( quartile range) 
    - extreme observations shown independently. 
    - Boxplot is plotted by setting kind=’box’ in catplot() function. 

```python
sns.catplot(data=auto_prices,x='body-style',y='engine-size',kind='box')
```

#### 2. Boxenplot

- Related function to box plot is boxenplot(), 
- It shows more information about distribution. 
- It is more suitable for large dataset.
- set kind='boxen' in catplot() function.


```python
sns.catplot(data=auto_prices,x='body-style',y='engine-size',kind='boxen')
```

#### 3. Violinplots 

- Based on hybrid approach that combine boxplot with kernel density estimation procedure. 
- So it represent more information on same plot. 
- Set kind=’violin’ in catplot() method of seaborn. 
- Thick bar in centre represents the inter-quartile range
- Thin black line extended from it represents the 95% confidence intervals
- White dot is the median.


```python
sns.catplot(data=auto_prices,x='body-style',y='engine-size',kind='violin')
```

## Visualizing the distribution of a dataset
- When dealing with data, it is an important task to know about how the variables are distributed.
- or to plot univariate and bivariate distribution of variables.


1. ### Plotting univariate distribution
      - The simplest way to plot a univariate distribution in seaborn by distplot(). 
      - By default it draw a histogram and kernel density estimation (KDE) on same plot.
      - pass the variable as an argument to distplot() function.
      - Another important argument is **bins**, by default it is 10.


```python
sns.distplot(auto_prices['engine-size'])
```

#### Histogram

- Diagram consisting of bins, along the range of variable.
- Area of bins are proportional to the number of observation in each bin. 
- To plot a histogram remove KDE curve in distplot() by setting kde=False.



```python
sns.distplot(auto_prices['engine-size'],kde=False)
```

#### Kernel density estimation(KDE)
- KDE is used to plot the shape of distribution. 
- It has observation on one axis and their density on other axis. 
- KDE is plot by setting hist=False in distplot() function of seaborn or using kdeplot() function directly.

```python
sns.distplot(auto_prices['engine-size'],hist=False)
```

### Plotting bivariate distributions
- jointplot() function is used to visualize bivariate distribution.
- It creates a multi-panel figure 
- that shows both the bivariate as well as univariate (or marginal) distributions on separate axes.

#### Scatter plots

- Scatter plot most common way to visualize a bivariate distribution.
- Each data point is shown as a dot of x and y values. 
- jointplot() function by default plot scatter plot.
- If some dots are over-plotted on each other
- it is difficult to know actually how many dots are plotted. 
- That create difficulty in visualization.


```python
sns.jointplot(auto_prices['price'],auto_prices['engine-size'])
```

#### Hexbin plots

- Hexbin plots are basically bivariate histograms.
- As it shows the number of observations that fall within hexagonal bins. 
- Suitable for large datasets, where over-plotting of points occurs. 
- As a general rule scatter plots are used when data is small 
- in case of large datasets, use hexbin plot. 
- Set kind=”hex” in jointplot() function or by hexbin() function directly.


```python
sns.jointplot(auto_prices['price'],auto_prices['engine-size'],kind='hexbin')
```

```python
auto_prices['price']=pd.to_numeric(auto_prices['price'])
```

#### Kernel density estimation
- It is also possible to use kernel density estimation to visualize a bivariate distribution. 
- this kind of plot is shown with a contour plot. 
- Set kind parameter to “kde” in distplot() function, or use kdeplot() function directly

```python
sns.jointplot(data=auto_prices,x='engine-size',y='price')
```

## Visualizing pairwise relationships in a dataset
- Multiple variables are to be plot on the same plot. 
- Already multiple variables are plotted in same figure. 
- But they were categorical variables. 
- For quantitative variables, pairplot() function is used. 
- This creates a matrix of axes and shows the relationship of each pair of columns 
- By default off diagonal axes are plotting a scatter plot and diagonal axes plot hist. 
- Diagonal plot can be changed to kde by setting diag_kind parameter to kde.


```python
sns.pairplot(data=auto_prices,vars=['city-mpg','highway-mpg', 'price'])
```

## Linear regression models:

- Linear relationship between two variables is determined through linear regression. 
- Seaborn has two functions, regplot() and lmplot(), to visualize linear relationships. 
- They both are closely related
- Both draw a scatterplot of two variables, x and y
- then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.
- pass both variables in this function as an x and y arguments. 
- By default confidence interval width is 95%
- but it can be changed by assigning integer value in [0,100] to argument “ci”.


```python
sns.regplot(data=auto_prices,x='engine-size',y='price')
```

### Plotting regression line using lmplot():
- Pass the x and y variables In lmplot()
- lmplot() function accepts hue, row, and col. 
- So lmplot() function can be extended to multi-dimensions.


```python
sns.lmplot(data=auto_prices,x='engine-size',y='price')
```

## Heatmap
- Heatmap uses a system of color-coding to represent different values. 
- A simple heat map provides an immediate visual summary of data. 
- More elaborate heat maps allows viewers to understand complex data sets.
- Before plotting heat map, cross-tabulated or pivoted data is required. 
- In seaborn, heatmap() function is used to plot a heat map.
- An important parameter of heatmap() function is “annot”.
- By default it is false, but when it is set to “True”, it will place the values in respective cells.



```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(figsize=(10,10))
auto_cross=pd.crosstab(columns=[auto_prices['body-style'],auto_prices['fuel-type']],
                      index=[auto_prices['num-of-cylinders'],auto_prices['drive-wheels']])
sns.heatmap(data=auto_cross)
```

## Dash interactivity
- Dash interactivity is possible due to the dash call-back decorators. 
- These decorators sensitive to the change in properties of components.



In this example above,”value” property of the Slider is the input of the app and the output of the app is the “figure” property of the Graph. Whenever the value of the Slider changes, Dash calls the callback function update_figure with the new value. The function filters the dataframe with this new value, constructs a figure object, and returns it to the Dash application.
There are a few nice patterns in this example:
1.	Pandas library for importing and filtering datasets in memory.
2.	Dataframe df is loaded at the start of the app. This dataframe df is in the global state of the app and can be read inside the callback functions.
3.	The callback does not modify the original data, it just creates copies of the dataframe by filtered through pandas filters. This is important that callbacks should never mutate variables outside of their scope. If callbacks modify global state, then one user's session might affect the next user's session and when the app is deployed on multiple processes or threads, those modifications will not be shared across sessions.
